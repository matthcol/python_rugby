```python
f = open('data/Test import json.json')
data = json.load(f)
# data
```

```python
data.keys()
```

```python
lignes = data['rows']
# lignes
```

```python
for ligne in lignes[:3]:
    print(ligne.keys())
```

```python
# ligne #2 : touches adverses
ligne = lignes[2]
name = ligne['row_name']
print("Nom d'action:", name)
print(ligne.keys())
clips = ligne['clips']
# clips
```

```python
len(clips)
```

```python
liste_actions = []
for touche in clips:
    for info in touche['qualifiers']['qualifiers_array']:
        if info['category'] == 'ZONES':
            zone = info['name']
        elif info['category'] == 'CONQUETE EFFICACITE':
            conquete_efficacite = info['name']
    # print("Zone:", zone)
    # print("Conquete efficacité:", conquete_efficacite)
    action = { 'zone': zone, 'conquete_effcicacite': conquete_efficacite }
    # print(action)
    liste_actions.append(action)
liste_actions
          
```

```python
dfTouchesAdverses = pd.DataFrame(liste_actions)
dfTouchesAdverses
```
